```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import sys
path_file = ('D:\\Laura_Rojas_Gonzalez\\SimpleSpace\\Datamanager')
sys.path.append(path_file)
from datamanager import DataManager
path_file = ('D:\\Laura_Rojas_Gonzalez\\SimpleSpace\\Calibrator')
sys.path.append(path_file)
from calibrator import Calibrator
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
import matplotlib.gridspec as gridspec
import matplotlib.ticker as ticker
from matplotlib.lines import Line2D
from PyPDF2 import PdfFileMerger
```

```python
path_general = 'D:\\Laura_Rojas_Gonzalez\\SimpleSpace\\La_Romera\\'
path_save = 'D:\\Laura_Rojas_Gonzalez\\SimpleSpace\\Reporte_MinCiencias'
#Obtenemos una lista del nombre de los archivos dentro de la carpeta
list_name = os.listdir(path_general)

#Iteramos para abrir cada archivo y concatenamos para tener un solo df

df_romera = []
for file in list_name:
    load_file = pd.read_csv(f'{path_general}{file}')
    df_romera.append(load_file)
df_romera = pd.concat(df_romera)
```

```python
df_romera.reset_index()
df_romera.reset_index(inplace=True, drop=True)
```

```python
df_romera
```

```python
df_romera['time']=df_romera['time'].apply(lambda x: dt.strptime(x, '%H:%M:%S'))
```

```python
for i in range(0,44):
    df_romera['time'][i]=df_romera['time'].iloc[i].replace(year=2023,month=4)
```

```python
for i in range(0,7):
    df_romera['time'][i]=df_romera['time'].iloc[i].replace(day=11)
```

```python
for i in range(7,37):
    df_romera['time'][i]=df_romera['time'].iloc[i].replace(day=12)
```

```python
for i in range(37,44):
    df_romera['time'][i]=df_romera['time'].iloc[i].replace(day=13)
```

```python
#Para cambiar los segundos a 00
for i in range(0,44):
    df_romera['time'].loc[i]=df_romera['time'].loc[i].replace(second=0)
```

```python
df_romera.index = df_romera['time']
df_romera.drop(['time'],axis=1,inplace=True)
```

```python
df_romera = df_romera.resample('T').mean()
```

```python
df_romera
```

```python
df_romera['barometer_pressure']=df_romera['barometer_pressure']/100
```

```python
#Creamos la instancia de la clase con el df cargado
Datamanager = DataManager(dataframe = df_romera)
#Cargamos los datos en un dataframe del datamanager
df_laromera = Datamanager.load_data()
```

```python
#Miramos la calidad de datos
Datamanager.show_period(df_laromera)
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111)
plt.plot(df_laromera.index,df_laromera.CO.values, lw = 5, marker = 'o',color='#056674')
ax.set_ylim(-5,5)
ax.set_ylabel(r'ppm', fontsize=15,  color='black')
ax.set_xlabel("Tiempo", fontsize=10, color='black')
# #ax.set_xlim(data.index[0], data.index[-1])
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M\n%b-%d'))
ax.tick_params(axis = 'y',color='black', labelcolor='black')
ax.tick_params(axis = 'x',color='black', labelcolor='black')
ax.grid(which='major',  linestyle='--', linewidth=0.5, color='lightgray')
ax.set_title('Serie CO', fontsize=20, color='gray')
# ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=8))

# plt.legend(loc= 0, fontsize = labelsize)

min_x = np.argmin(df_laromera['CO'])
min_y = np.min(df_laromera['CO'])
min_y = round(min_y,2)

max_x = np.argmax(df_laromera['CO'])
max_y = np.max(df_laromera['CO'])
max_y = round(max_y,2)

plt.scatter(min_x, min_y,c='k')
plt.scatter(max_x, max_y,c='r')

plt.annotate(min_y,(min_x,min_y),textcoords="offset points",xytext=(0,-10),ha='center')
plt.annotate(max_y,(max_x,max_y),textcoords="offset points",xytext=(0,5),ha='center')

plt.axhline(y=np.nanmean(df_laromera['CO']),color='gray', lw=1.5, ls='--', label="promedio")
plt.legend(loc='lower left')

plt.show()
```
